## Getting the dataset

```python
# For data wrangling 
import numpy as np
import pandas as pd

# For visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
pd.options.display.max_rows = None
pd.options.display.max_columns = None
```

```python
from datetime import datetime
import math

df1 = pd.read_csv(r"../data/train.csv").drop(columns=['washDishes'])
df2 = pd.read_csv(r"../data/test.csv")

df1['key'] = 'train'
df2['key'] = 'test'
```

```python
labels = 'Pet', 'no Pet'
sizes1 = [df1.hasPet[df1['hasPet']==1].count(), df1.hasPet[df1['hasPet']==0].count()]
sizes2 = [df2.hasPet[df2['hasPet']==1].count(), df2.hasPet[df2['hasPet']==0].count()]

explode = (0, 0.1)

fig1, ax1 = plt.subplots(nrows=1, ncols=2, figsize=(20, 16))

ax1[0].pie(sizes1, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1[0].set_title('Train', size=20)

ax1[1].pie(sizes2, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1[1].set_title('Test', size=20)

# plt.title("Gender Distribution", size = 30)
plt.show()
```
