## Getting the dataset

```python
## REQUIRED LIBRARIES
# For data wrangling 
import numpy as np
import pandas as pd

# For visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
pd.options.display.max_rows = None
pd.options.display.max_columns = None
```

```python
from datetime import datetime
import math

df1 = pd.read_csv(r"data/train.csv").drop(columns=['labels'])
df2 = pd.read_csv(r"data/test.csv")

df1['key'] = 'train'
df2['key'] = 'test'

df = pd.concat([df1,df2],keys=['train','test'])
```

```python
## Relations
```

```python
# We first review the relation of automatic dishwasher owners with categorical variables
fig, axarr = plt.subplots(22, 1, figsize=(30, 100))
for i, c in enumerate(df.columns):
    if 1 < i < 23: sns.boxplot(x='key', y=c, data = df, ax=axarr[i - 1])
```
